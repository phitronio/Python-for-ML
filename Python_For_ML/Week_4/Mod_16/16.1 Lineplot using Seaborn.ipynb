```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from matplotlib.pyplot import ylabel
```

# **_16.1 Lineplot using Seaborn_**

```python
student = pd.read_csv('sns_data.csv')
student.head()
```

```python
sns.lineplot(data=student, x ='week', y = 'attendance_rate',errorbar=None
             ,hue='gender')
```

```python
sns.relplot(kind='line',data=student, x ='week', y = 'attendance_rate',errorbar=None
             ,hue='gender')
```

```python
sns.relplot(kind='line',data=student, x ='week', y = 'attendance_rate',errorbar=None
             ,hue='class_level')
```

# **_16.2 Scatterplot using Seaborn_**

```python
sns.scatterplot(data=student, x='study_hours', y='test_score',hue ='gender',style='class_level', size='Tshirt_size')
```

```python
tips_data = sns.load_dataset("tips")

tips_data.head()
```

```python
sns.relplot(kind='scatter', data =tips_data, x = 'total_bill', y = 'tip', hue='sex',
            style='time',size='size')
```

# **_16.3 Facet Plot Using Seaborn_**

```python
student = pd.read_csv('student_dataset_complete.csv')
student.head()
```

```python
sns.scatterplot(data= student, x = 'study_hours', y='test_score', hue='gender')
```

```python
sns.relplot(kind='scatter',data= student, x = 'study_hours', y='test_score', hue='gender', col='gender', row='hostel')
```

```python
sns.relplot(kind='scatter',data= student, x = 'study_hours', y='test_score', hue='gender', col='week',col_wrap=2)
```

# **_16.4 Histogram using Seaborn_**


```python
# axes level
sns.histplot(data = student , x ='attendance_rate',hue='gender',bins=15,element='step')
```

```python
sns.displot(kind='hist', data = student, x = 'study_hours',col='gender',element='step')
```

# **_16.5 KDE plot using Seaborn_**

```python
sns.displot(kind='hist', data = tips_data, x = 'total_bill', bins=20)
```

```python
sns.displot(kind='hist', data = tips_data, x = 'tip', bins=20)
```

```python
# axes level
sns.kdeplot(data = student , x ='attendance_rate')
```

```python
sns.kdeplot(data= tips_data, x ='tip', hue='sex', fill=True)
```

# **_16.6 Count Plot and Bar Plot using Seaborn_**


```python
sns.countplot(data=student, x ='gender')
```

```python
sns.countplot(data=student, x ='subject', hue='gender')
```

```python
sns.barplot(data=student, x ='gender', y='Marks', errorbar=None)
```

```python
sns.barplot(data=student, x ='gender', y='Marks', errorbar=None, estimator=np.max)
```

# **_16.7 Regplot using seaborn_**


```python
sns.regplot(data =student, x ='study_hours', y= 'test_score')
```

```python
sns.lmplot(data =student, x ='study_hours', y= 'test_score', hue='gender')
```

# **_16.8 Pairplot and Jointplot using Seaborn_**

```python
student_marks = student[['Marks','study_hours','attendance_rate','gender']]
student_marks
```

```python
sns.pairplot(data=student_marks)
```

```python
sns.pairplot(data=student_marks,kind='hist')
```

```python
sns.jointplot(data=student_marks, x='study_hours', y='Marks')
```

```python
sns.jointplot(data=student_marks, x='study_hours', y='Marks', kind='kde')
```

```python
sns.jointplot(data=student_marks, x='study_hours', y='Marks', hue='gender')
```
