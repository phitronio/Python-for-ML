```python
import seaborn as sns
import pandas as pd

# Load data
student = pd.read_csv('sns_data.csv')

# Create line plot
sns.lineplot(data=student, x='week', y='attendance_rate', hue='gender', errorbar=None)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Using the student dataset
sns.scatterplot(data=student, x='study_hours', y='test_score', hue='gender', size='attendance_rate')

plt.title("Relationship: Study Hours vs Test Score")
plt.xlabel("Study Hours")
plt.ylabel("Test Score")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
student = pd.read_csv('sns_data.csv')

# Create scatter plot
sns.scatterplot(
    data=student,
    x='study_hours',
    y='test_score',
    hue='gender',
    style='class_level',
    size='attendance_rate',
    alpha=0.7
)

# Customize
plt.title("Study Hours vs Test Score")
plt.xlabel("Study Hours")
plt.ylabel("Test Score")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram of test scores
sns.histplot(data=student, x='test_score', bins=15, kde=True, color='skyblue')

plt.title("Distribution of Test Scores")
plt.xlabel("Test Score")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
student = pd.read_csv('sns_data.csv')

# Create histogram
sns.histplot(
    data=student,
    x='test_score',
    bins=20,
    kde=True,           # Adds smooth density curve
    color='lightgreen',
    alpha=0.7
)

# Customize
plt.title("Distribution of Test Scores")
plt.xlabel("Test Score")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# KDE plot of test scores
sns.kdeplot(data=student, x='test_score', hue='gender', fill=True, alpha=0.5)

plt.title("KDE: Distribution of Test Scores by Gender")
plt.xlabel("Test Score")
plt.ylabel("Density")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
student = pd.read_csv('sns_data.csv')

# Create KDE plot
sns.kdeplot(
    data=student,
    x='test_score',
    hue='gender',
    fill=True,          # Fill under the curve
    alpha=0.6,
    linewidth=2
)

# Customize
plt.title("KDE Plot: Test Score Distribution by Gender")
plt.xlabel("Test Score")
plt.ylabel("Density")
plt.legend(title="Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
student = pd.read_csv('sns_data.csv')

# Scatter plot with regression line
sns.regplot(
    data=student,
    x='study_hours',
    y='test_score',
    scatter_kws={'alpha': 0.6, 'color': 'blue'},
    line_kws={'color': 'red', 'linewidth': 2}
)

# Customize
plt.title("Study Hours vs Test Score (with Regression Line)")
plt.xlabel("Study Hours")
plt.ylabel("Test Score")
plt.show()
```
