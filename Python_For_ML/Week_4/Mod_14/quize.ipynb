```python
#3
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Data Structure Marks": [95, 85, 91]
})

df['Grade'] = np.where(df['Data Structure Marks'] > 90, 'A+', 'A')
print(df)
```

```python
#7
import pandas as pd

df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [10, 20, 30]
})

# প্রতি রো-তে যোগফল বের করি
df['Sum'] = df.apply(lambda row: row['A'] + row['B'], axis=1)

print(df)

# প্রতি রো-তে "A > B?" চেক করি
df['A_gt_B'] = df.apply(lambda row: row['A'] > row['B'], axis=1)
# আউটপুট: [False, False, False]
```

```python
#8
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob"],
    "EnrollmentDate": ["2024-01-15", "2023-06-20"]
})

# প্রথমে datetime-এ কনভার্ট করতে হবে
df['EnrollmentDate'] = pd.to_datetime(df['EnrollmentDate'])

# এখন .dt.year ব্যবহার করি
df['Year'] = df['EnrollmentDate'].dt.year

print(df)
```

```python
#9
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["A", "B", "C", "D"],
    "Python Marks": [85, np.nan, 92, np.nan]
})

print("আগে:")
print(df)

df['Python Marks'].fillna(df['Python Marks'].mean(), inplace=True)

print("\nপরে:")
print(df)
```
