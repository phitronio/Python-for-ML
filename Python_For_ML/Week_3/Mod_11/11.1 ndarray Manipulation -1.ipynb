
#                                          **11.3 ndarray Arithmetic Operators and Mathematical function**

```python
import numpy as np

x1 = np.array([ 10, 20, 30, 100])
y1 = np.array ([2, 3, 4, 5])

add = x1+y1
print(add)

sub = x1-y1
print(sub)

div = x1/y1
print(div)

remainder = x1 % y1
print(remainder)
```

```python
sin_val = np.sin(x1)
print(sin_val)

cos_val = np.cos(x1)
print(cos_val)
print()
base_10_log = np.log10(x1)
print(base_10_log)
```

```python
ase_10_log = np.log2(x1)
print(base_10_log)
```

```python
sqr = np.sqrt(x1)
print(sqr)
```

```python
s = np.sum(x1)
print(s)
cumul =np.cumulative_sum(x1)
print(cumul)
```
