```python
import numpy as np

# Define the matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
result = np.dot(A, B)

# Print the result
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nResult of dot(A, B):")
print(result)
```
