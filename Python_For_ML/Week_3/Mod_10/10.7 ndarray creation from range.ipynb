```python
import numpy as np

arr = np.arange(1 , 10,1).reshape(3,3)

print(arr)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
import numpy as np

arr = np.linspace(1 , 10,15)

print(arr)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
arr = np.logspace(1 , 10,15)

print(arr)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
arr = np.logspace(0,4,6, base =2)

print(arr)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```
