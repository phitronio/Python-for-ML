```python
import numpy as np

diagonal_matrix = np.diag([1,2,3,4])

print(diagonal_matrix)
print(diagonal_matrix.shape)
```

```python
# identity matrix

identity_mat = np.eye(4)
print(identity_mat)
print(identity_mat.shape)

```

```python

# np.eye(row , column,k)

mat = np.eye(3,4)
print(mat)

mat = np.eye(3,4,1)
print(mat)
```
