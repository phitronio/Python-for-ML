# **12.5 Dataframe from Inbuilt Data with Column and Indexes**

```python
my_list = [['Alice', 25], ['Bob', 30], ['Charlie', 28]]

list_df = pd.DataFrame(my_list,columns=['Name','Age'],index=[1,2,3])

print(type(list_df))

list_df



```

```python
my_tuple = (('Alice', 25), ('Bob', 30), ('Charlie', 28))

tuple_df = pd.DataFrame(my_tuple)
tuple_df
```

```python
my_dict = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 29,28]
}

dict_df = pd.DataFrame(my_dict)

dict_df
```
